## Histogram for Aggregate Rating

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Aggregate rating'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()
```

## Countplot for Cuisines

```python
# 2. Countplot for Top 10 Cuisines
plt.figure(figsize=(8, 5))
cuisine_counts = df['Cuisines'].str.split(', ').explode().value_counts().head(10)
sns.barplot(x=cuisine_counts.index, y=cuisine_counts.values, palette='muted')
plt.title('Top 10 Cuisines')
plt.xlabel('Cuisine Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## Scatterplot: Aggregate Rating vs Votes

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='Aggregate rating', y='Votes',hue='Price range',color='viridis')
plt.title("Aggregate rating vs Votes")
plt.xlabel("Aggregate rating")
plt.ylabel("Votes")
plt.show()
```

## Boxplot for Outliers in Votes

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df,x='Votes',color='lightblue')
plt.title("Outliers in Votes")
plt.xlabel('Votes')
plt.show()
```

##  Heatmap for Correlations

```python
correlation_matrix = df[['Aggregate rating', 'Votes', 'Price range']].corr()
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
